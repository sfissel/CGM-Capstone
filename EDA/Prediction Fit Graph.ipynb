```python
import pandas as pd
import polars as pl
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Model_Predictions.csv') 
```

```python
def plot_reading(df, i):
    start_index = i * 288
    end_index = (i + 1) * 288
    df1=df[start_index:end_index]
    real_data = df1['Value_Complete']
    missing_data_mask = df1['Value_Missing'].isna()  


    plt.figure()
    plt.plot(df1.DailyTI_Index, real_data, label='Real Data', color='black')


    for column in df1.columns:
        if 'predictions' in column:  

            imputed_data = df1[column]

            plt.plot(df1.DailyTI_Index[missing_data_mask], imputed_data[missing_data_mask], 
                 label=f'Imputed with {column}',  marker='o', linestyle='-', markersize=2, alpha=0.5)

    plt.xlabel('Index')
    plt.ylabel('Values')
    plt.title('Real vs Imputed Values')
    plt.legend()
    plt.show()
```

```python
plot_reading(df,55)
```
